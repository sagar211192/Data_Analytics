```python
import h2o
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
h2o.init()
```

```python
train = h2o.upload_file("F:\DA\Dissertation\TFIDF_train_data - Copy.csv")
test = h2o.upload_file("F:\DA\Dissertation\TFIDF_test_data - Copy.csv")
```

```python
y = "airline_sentiment"
x = train.names[0:3640]
```

```python
train[y] = train[y].asfactor()
test[y] = test[y].asfactor()
```

```python
ber_model = H2ODeepLearningEstimator(distribution="bernoulli", activation="Tanh", hidden=[32, 20,34],
                                 input_dropout_ratio=0.2, max_confusion_matrix_size= 2, sparse=True, l1=1e-5, epochs=10)
ber_model.train(x=x, y=y, training_frame=train, validation_frame=test)
```

```python
prediction = ber_model.predict(test)
prediction
```
